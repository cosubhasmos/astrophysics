### Methodology and Code

To start up with plotting I have just used two popular python libraries <i>Pandas</i> for data handling and <i>Matplotlib</i> for plotting. We can just start up with importing them as:

```python
import matplotlib.pyplot as plt
import pandas as pd
```

Now we can read our datasets using <code>pd.read_csv</code> function. Since, the dataset only contains two columns, Wavelength and Intensity, we can specify them as headers. 

```python
data1 = pd.read_csv("sn2023_specturm_ascii_file.dat",delimiter=" ",header=None,names=['Wavelength', 'Intensity'])
data2 = pd.read_csv("tns_2023ixf_2023-05-22_23-50-00_Other_Other.dat",delimiter=' ',header=None,names=['Wavelength', 'Intensity'],dtype={'Wavelength': str})
data3 = pd.read_csv("tns_2023ixf_2460087.272_HCT-2m_HFOSC.ascii",sep=r'\s+',header=None,names=['Wavelength', 'Intensity'])
data4 = pd.read_csv("tns_2023ixf_2460085.378_HCT-2m_HFOSC.ascii",sep=r'\s+',header=None,names=['Wavelength', 'Intensity'])
data5 = pd.read_csv("tns_2023ixf_2023-05-25_06-42-58_P60_SEDM_ZTF.txt",comment='#',delimiter=' ',header=None,names=['Wavelength', 'Intensity','Intensity2'])
data6 = pd.read_csv("tns_2023ixf_2023-05-10_22-23-45.175_LT_SPRAT.spec",sep=r'\s+',comment='#',header=None,names=['Wavelength', 'Intensity','sky'])
```

Now we can just arrange those datasets and corresponding names in array which will make plotting easier.

```python
datasets = [data1, data2, data3,data4,data5,data6]
names = ["SN_2023ixf - 2023-07-21 21:44:00 Other / Other","tns_2023ixf_2023-05-22_23-50-00_Other_Other","tns_2023ixf_2460087.272_HCT-2m_HFOSC","tns_2023ixf_2460085.378_HCT-2m_HFOSC","tns_2023ixf_2023-05-25_06-42-58_P60_SEDM_ZTF","tns_2023ixf_2023-05-10_22-23-45.175_LT_SPRAT"]
color = ["red","green","blue","orange","violet","purple"]
```

Now we can run a loop times the length of dataset array and make corresponding plots for each datasets.

```python
fig, axs = plt.subplots(len(datasets),1,figsize=(20,30))

for i, df in enumerate(datasets):
    ax = axs[i]
    x = df["Wavelength"]
    y = df["Intensity"]
    ax.plot(y,linewidth=3,color=color[i],label=names[i])
    ax.legend(loc="upper right",fontsize=26)
ax.set_xlabel("Wavelength",fontsize=20)
fig.text(0, 0.5, 'Intensity', va='center', rotation='vertical', fontsize=20)
fig.tight_layout()
plt.show()
```
